```python
# import libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

## Read & Inspecting the data

```python
df=pd.read_csv('World Happiness Report 2022.csv')
```

```python
df.columns
```

```python
df=df.rename(columns={'Explained by: GDP per capita':'GDP per capita',
            'Explained by: Social support':'Social support', 
            'Explained by: Healthy life expectancy':'Healthy life expectancy',
            'Explained by: Freedom to make life choices':'Freedom to make life choices', 
            'Explained by: Generosity':'Generosity',
            'Explained by: Perceptions of corruption':'Perceptions of corruption'})
df.head()
```

```python
df.describe().style.background_gradient()
```

## Top 10 Countries

```python
top= df.sort_values('Happiness score', ascending=False)[:10].sort_values('Happiness score')
fig= px.bar(top, x='Happiness score', y='Country',color='RANK', height=800, title='Top 10 countries')
fig.show()
```

## Last 10 countries

```python
last= df.sort_values('Happiness score')[:10].sort_values('Happiness score', ascending=False)
fig= px.bar(last, x='Happiness score', y='Country', height=800, title='Last 10 countries')
fig.show()
```

```python
# Relationship
plt.figure(figsize=(11,6))
sns.heatmap(df.corr(), annot=True)
```

```python
data=df.select_dtypes(include=[np.number])
def hist_plot(var, data):
    sns.distplot(data[var])
    plt.axvline(x=data[var].mean(), color='r')
    plt.axvline(x=data[var].median(), color='b')
    plt.title(f'{var}')
    plt.show()

for var in data:
    hist_plot(var, data)
```

```python
# GDP of top 10 countries
plt.figure(figsize=(9,9))
data=df.sort_values('GDP per capita',ascending=False)[:10]
fig=px.pie(data, values='GDP per capita',
           names='Country',title='GDP per capita of top 10 countries')
fig.show()
```

## Relation of Happiness Score with factors

```python
fig=px.scatter(df, y='Happiness score', x='GDP per capita',size='Happiness score',
               title='Happiness score Vs GDP')
fig.show()
```

```python
fig=px.scatter(df,x='Social support', y='Happiness score', 
               size='Happiness score',color='Country',
               title='Happiness score Vs Social support')
fig.show()
```

```python
plt.figure(figsize=(9,9))
fig=px.scatter(df,x='Healthy life expectancy', y='Happiness score', 
               hover_data=['Country'],color='RANK',
               title='Happiness score Vs Healthy life expectancy')
fig.show()
```

```python
fig=px.scatter(df,x='Freedom to make life choices', y='Happiness score', 
               size='Happiness score',color='RANK',
               title='Happiness score Vs Freedom to make life choices')
fig.show()
```

```python
# Black spot mention Pakistan
fig=px.scatter(df,x='Perceptions of corruption', y='Happiness score', 
               size='Happiness score',color='Country',color_discrete_map={'Pakistan':"black"},
               
               title='Happiness score Vs Perceptions of corruption')
fig.show()
```

```python
# Black spot mention Pakistan
fig=px.scatter(df,x='Perceptions of corruption', y='GDP per capita', 
               size='Happiness score',color='Country',color_discrete_map={'Pakistan':"black"},
               
               title='GDP per capita Vs Perceptions of corruption')
fig.show()
```

```python
top=df.sort_values(by='Happiness score', ascending=False).head(10)
tail=df.sort_values(by='Happiness score').head(10)
f,(ax1,ax2)=plt.subplots(1,2,figsize=(15,15))
top.plot(kind='barh',x='Country',y='Happiness score'
                    ,ax=ax1,color='blue', title='top 10 countries')
tail.plot(kind='barh',x='Country',y='Happiness score'
                    ,ax=ax2,color='green',title='Last 10 countries' )
```

```python
score=df.sort_values(by='Happiness score', ascending=False).head(10)
gdp=df.sort_values(by='Happiness score', ascending=False).head(10)
f,(ax1,ax2)=plt.subplots(1,2,figsize=(15,10))
score.plot(kind='barh',x='Country',y='Happiness score'
                    ,ax=ax1,color='red', title='Happiness score rank')
gdp.plot(kind='barh',x='Country',y='GDP per capita'
                    ,ax=ax2,color='green', title='GDP rank')
```
